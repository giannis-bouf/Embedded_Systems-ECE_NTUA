```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Άσκηση 1

# Δεδομένα
data_str = ["(SLL,SLL)", "(SLL,DLL)", "(SLL,DYN_ARR)",
            "(DLL,SLL)", "(DLL,DLL)", "(DLL,DYN_ARR)",
            "(DYN_ARR,SLL)", "(DYN_ARR,DLL)", "(DYN_ARR,DYN_ARR)"]
mem_access = np.array([71817154, 72488797, 470628564,
                       71829818, 72500828, 470643127,
                       72294688, 72982710, 471361860])
mem_footprint = np.array([798.8, 980.3, 1111, 823, 983.3, 1128, 760.2, 982.5, 1075])

# Δημιουργία subplots
fig, axs = plt.subplots(2, 1, figsize=(15, 8))

# Πρώτο subplot
axs[0].plot(data_str, mem_access, marker='o', color='navy')
axs[0].set_title("Memory Accesses vs Data Structure")
axs[0].set_xlabel("Data Structure (clientList, pList)")
axs[0].set_ylabel("Memory Accesses")
axs[0].set_xticks(data_str)
axs[0].grid()

# Δεύτερο subplot
axs[1].plot(data_str, mem_footprint, marker='o', color='navy')
axs[1].set_title("Memory Footprint vs Data Structure")
axs[1].set_xlabel("Data Structure (clientList, pList)")
axs[1].set_ylabel("Memory Footprint (kB)")
axs[1].set_xticks(data_str)
axs[1].grid()

plt.tight_layout()
plt.show()
```

```python
# Άσκηση 2

files = ["sll.txt", "dll.txt", "dyn_arr.txt"]

with open("original.txt", 'r', encoding='utf-8') as original:
    original_content = original.read()
    for f in files:
        with open(f, 'r', encoding='utf-8') as new_file:
            content = new_file.read()

        # Σύγκριση περιεχομένου
        if original_content == content:
            print(f"'original.txt' and '{f}' have the same content!")
        else:
            print(f"'original.txt' and '{f}' have different content!")
```
