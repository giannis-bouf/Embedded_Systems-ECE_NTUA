```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1.2
no_op = np.array([0.00204799999, 0.00204800931, 0.00204800931, 0.00204800931, 0.00204800931,
                  0.00204800931, 0.00204800931, 0.00204800931, 0.00204800931, 0.00204800931])

print(f"Μέσος Όρος: {np.mean(no_op)} sec")
print(f"Μέγιστη Τιμή: {np.max(no_op)} sec")
print(f"Ελάχιστη Τιμή: {np.min(no_op)} sec")

plt.plot(no_op, linestyle='--', label='no_op')
plt.xlabel("Iteration")
plt.ylabel("Seconds")
plt.ylim(0.00204775, 0.00204825)
plt.grid()
plt.show()
```

```python
# 1.3
inter = np.array([0.00205016672, 0.00205017603, 0.00205017603, 0.00205017603, 0.00205017603,
                  0.00205017603, 0.00205017603, 0.00205017603, 0.00205017603, 0.00205017603])
merging = np.array([0.0018434861,  0.00184349541, 0.00184349541, 0.00184349541, 0.00184349541,
                    0.00184349541, 0.00184349541, 0.00184349541, 0.00184349541, 0.00184349541])
merg_int = np.array([0.0018429769,  0.00184298609, 0.00184298609, 0.00184298609, 0.00184298609,
                     0.00184298609, 0.00184298609, 0.00184298609, 0.00184298609, 0.00184298609])
pre_calc = np.array([0.00183160184, 0.00183157402, 0.00183157402, 0.00183157402, 0.00183157402,
                     0.00183157402, 0.00183157402, 0.00183157402, 0.00183157402, 0.00183157402])

mean = np.array([np.mean(no_op), np.mean(inter), np.mean(merging), np.mean(merg_int), np.mean(pre_calc)])
opt = ["no_op", "inter", "merging", "merg_int", "pre_calc"]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(no_op, linestyle='--', label='no_op')
plt.plot(inter, linestyle='--', label='inter')
plt.plot(merging, linestyle='--', label='merging')
plt.plot(merg_int, linestyle='--', label='merg_int')
plt.plot(pre_calc, linestyle='--', label='pre_calc')
plt.xlabel("Iteration")
plt.ylabel("Seconds")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(opt, mean, marker='o', color='purple')
plt.title("Average Time")
plt.xlabel("Optimization")
plt.ylabel("Seconds")
plt.xticks(opt)
plt.grid()

plt.tight_layout()
plt.show()
```
